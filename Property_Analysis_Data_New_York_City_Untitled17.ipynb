```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Load the uploaded CSV file (it's a CSV, not an Excel file)
file_path = "/content/property_analysis_data_new_york_city.csv"

# Use pd.read_csv to read CSV files
try:
    df = pd.read_csv(file_path)
    # Since CSV files don't have sheet names, we'll print the column names instead
    print("Column names:", df.columns.tolist())
except FileNotFoundError:
    print(f"Error: File not found at path: {file_path}")
except pd.errors.ParserError:
    print(f"Error: Unable to parse the CSV file. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

**The dataset has been successfully loaded and contains the following columns relevant to Airbnb listings in New York City:**

* **neighbourhood_group:** Indicates the major borough (e.g., Manhattan, Brooklyn, Queens).

* **neighbourhood:** Specific neighborhood within the borough.

* **price, number_of_reviews, availability_365,** etc.

```python
# Display the first few rows and column names to understand the structure
df.head(), df.columns
```

```python
# Use Case 1: Manhattan Neighbourhood
# Select subset for Manhattan
manhattan_df = df[df['neighbourhood_group'] == 'Manhattan']

# Number of rows
manhattan_rows = manhattan_df.shape[0]

# Number of columns
manhattan_columns = manhattan_df.shape[1]

(manhattan_df.shape[0], manhattan_df.shape[1])
```

# **Subset Information:**
* Total rows (listings) in Manhattan: 21,661
* Total columns: 16



```python
# Histogram of prices
plt.figure(figsize=(10, 5))
sns.histplot(manhattan_df['price'], bins=50, kde=True)
plt.title('Distribution of Prices in Manhattan (Airbnb)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.xlim(0, 1000)  # Trim outliers for better visibility
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter plot of number of reviews vs. price
plt.figure(figsize=(10, 5))
sns.scatterplot(data=manhattan_df, x='price', y='number_of_reviews', alpha=0.6)
plt.title('Number of Reviews vs Price in Manhattan')
plt.xlabel('Price ($)')
plt.ylabel('Number of Reviews')
plt.xlim(0, 1000)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Visual Insights:**
**Price Distribution (Histogram):**
* Most Airbnb listings in Manhattan are priced below 500 Dollars.
* A large concentration of prices falls between 50 Dollars and 300 Dollars.
* There are some high-end outliers, which were trimmed in the plot for clarity.

**Reviews vs. Price (Scatter Plot):**
* Listings with lower prices generally receive more reviews.
* There’s no strong linear relationship between price and review count.
* High-priced listings usually have fewer reviews.


```python
# Top 10 neighbourhoods in Manhattan by count
top10_manhattan = manhattan_df['neighbourhood'].value_counts().head(10)

# Subset of top 10 neighbourhoods
top10_manhattan_df = manhattan_df[manhattan_df['neighbourhood'].isin(top10_manhattan.index)]

# Number of rows
top10_manhattan_rows = top10_manhattan_df.shape[0]

# Number of columns
top10_manhattan_columns = top10_manhattan_df.shape[1]

# Count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=top10_manhattan_df, x='neighbourhood', order=top10_manhattan.index)
plt.title('Top 10 Manhattan Neighbourhoods by Airbnb Listings')
plt.xlabel('Neighbourhood')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

(manhattan_rows, manhattan_columns, top10_manhattan.index.tolist(), top10_manhattan_rows, top10_manhattan_columns)
```
